```python
import numpy as np
import pandas as pd
import difflib       # used to match the user input data , finding the closest correct match
from sklearn.feature_extraction.text import TfidfVectorizer  # used to convert the texctual data into numeric
from sklearn.metrics.pairwise import cosine_similarity  # need to find similarity score that's why we use this 
```

# Data collection and pre-processing

```python
movies_data = pd.read_csv("movies (1).csv")
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# taking selected columns
selected_features = ["genres", "keywords", "tagline", "cast", "director"]
print(selected_features)
```

```python
# Replacing the null values with null string

for i in selected_features:
    movies_data[i] = movies_data[i].fillna('')
```

```python
# comnining all the 5 selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
combined_features
```

```python
# Converting te text data to features vectors

Vectorizer = TfidfVectorizer()
```

```python
feature_vectors = Vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

# Cosine similarity

```python
#getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# getting the movie name from the user

Movie_name = input('Enter your favourite movie name : ')
```

```python
# creating the list with all the movies names given in the dataset

list_of_all_title = movies_data['title'].tolist()
```

```python
print(list_of_all_title)
```

```python
# finding the close match for the movie name given by the user

Finding_close_match = difflib.get_close_matches(Movie_name, list_of_all_title)
```

```python
print(Finding_close_match)
```

```python
close_match = Finding_close_match[0]
print(close_match)
```

```python
# findng the index of the movie with title

index_of_the_movie = movies_data[movies_data.title ==  close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))     #this is lower similarity score
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movie based on their similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i<21):
        print(i, '.', title_from_index)
        i+=1
```

# Movies recommendation system

```python
Movie_name = input('Enter your favourite movie name : ')

list_of_all_title = movies_data['title'].tolist()

Finding_close_match = difflib.get_close_matches(Movie_name, list_of_all_title)

close_match = Finding_close_match[0]

index_of_the_movie = movies_data[movies_data.title ==  close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse = True)

print('Movies suggested for you : \n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i<21):
        print(i, '.', title_from_index)
        i+=1
```
